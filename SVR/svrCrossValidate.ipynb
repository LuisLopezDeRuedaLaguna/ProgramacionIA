```python
from sklearn.svm import SVR
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate

```

```python
'''
kernel: 'linear', 'poly', 'rbf' o 'sigmoid' 
degree: solo para kernel 'poly' Cuantas curvas tiene el polinomio, 1 es lineal
gamma: para rbf y sifmoid: 'scale', 'auto' o float. Como de ancho y alto es, a mas alto a mas menos ancho
C: Lo que pagas por los fallos Para todos
epsilon: Si sale fuera la tolerancia que tiene para que acepte el punto Para todosss
'''

modelo = SVR(kernel="rbf",C=10, epsilon=0.5,gamma=10)
#modelo = SVR(kernel="linear",C=200, epsilon=1)
#modelo = SVR(kernel="poly",C=100, epsilon=0.5, degree=3)
#modelo = SVR(kernel="sigmoid",C=5   , epsilon=0.1,gamma=60)

data = datasets.load_diabetes()

X = data['data']
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
 
modelo.fit(X_train,y_train)

print(modelo.score(X_train,y_train))
print(modelo.score(X_test,y_test))


```

Cross validate, sirve para acotar los datos, en vez de salir muchos datos diferentes pues te lo deja en 1

```python
modelo = SVR(kernel="rbf",C=100, epsilon=0.5,gamma=10)
data = datasets.load_diabetes()

X = data['data']
y = data['target']
#kf = KFold(n_splits=5) NO hace falta porque viene en el cross_validate
#cross_validate(modelo,X,y, cv = kf)
dic = cross_validate(modelo,X,y, cv = 5, return_train_score=True) #Si no se pone nada coge kfol 5
print(dic['train_score'].mean())
print(dic['test_score'].mean())


```

```python
data = datasets.load_diabetes()

X = data['data']
y = data['target']
#kf = KFold(n_splits=5) NO hace falta porque viene en el cross_validate
#cross_validate(modelo,X,y, cv = kf)


valor = 0.1
nuevaC = 0
for c in [1,10,100,1000]:
    modelo = SVR(kernel="rbf",C=c, epsilon=0.5,gamma=10)
    dic = cross_validate(modelo,X,y, cv = 5, return_train_score=True)
    score_actual = dic['test_score'].mean()
    if score_actual > valor:
        valor = score_actual
        nuevaC = c



 #Si no se pone nada coge kfol 5
print(dic['train_score'].mean())
print(dic['test_score'].mean())     
print(valor, nuevaC)
```
