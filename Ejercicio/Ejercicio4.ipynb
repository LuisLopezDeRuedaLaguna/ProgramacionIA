```python
import numpy as np

a = np.arange(1, 100)

a1 = a.copy()    
np.random.shuffle(a1)
carton1 = a1[:10]

a2 = a.copy()   
np.random.shuffle(a2)
carton2 = a2[:10]

print("carton1 inicial:", carton1)
print("carton2 inicial:", carton2)

np.random.shuffle(a)
print(a)
i = 0

while carton1.size > 0 and carton2.size > 0 and i < a.size:
    numero = a[i]
    if numero in carton1:
        borrar = np.where(carton1 == numero)[0]
        carton1 = np.delete(carton1, borrar)
    if numero in carton2:
        borrar = np.where(carton2 == numero)[0]
        carton2 = np.delete(carton2, borrar)
    i += 1
print("Carton1 final:", carton1)
print("Carton2 final:", carton2)
print("NÃºmeros que quedan: ", a[i:])
```
