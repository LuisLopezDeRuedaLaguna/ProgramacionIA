```python
import numpy as np
import matplotlib.pyplot as plt

mapeo = {1: 'Piedra', 2: 'Papel', 3: 'Tijera'}
diccionario = {'Piedra': 0, 'Papel': 0, 'Tijera': 0}

tiradas_numeros = np.random.randint(1, 4, size=1000)

valores, cuentas_nuevas = np.unique(tiradas_numeros, return_counts=True)

for valor, cuenta in zip(valores, cuentas_nuevas):
    llave = mapeo[valor]
    diccionario[llave] += cuenta

print(f"Conteos finales: {diccionario}")

nombres = list(diccionario.keys())
conteos = list(diccionario.values())
print('Mximo: ', max(conteos), ' Mnimo: ', min(conteos))

plt.figure(figsize=(8, 5))
plt.bar(nombres, conteos, color=['#FFC300', '#C70039', '#581845'])

plt.title('Resultados de 49 tiradas (Solo NumPy y Matplotlib)')
plt.xlabel('Opcin elegida')
plt.ylabel('Frecuencia (Cantidad)')

plt.show()
```
