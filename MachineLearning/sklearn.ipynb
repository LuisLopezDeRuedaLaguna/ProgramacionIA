```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python

# #Carga de datos
x = np.arange(40,150,1)
datos = np.load('pisos_vendidos.npy')
m2 = datos[:,0]
precio = datos [:,1]

```

```python
# #Para sacar m y b 
# # M = modelo.coef_
# # B = modelo.intercept_

modelo = LinearRegression()
```

```python
#Hay que hacer una lista por cada 
X = m2.reshape(-1,1)
y = precio
modelo.fit(X,y)
print(f"M: {round(modelo.coef_[0],3)}, B: {round(modelo.intercept_,3)}")

plt.scatter(X,y)

y_pred = X*modelo.coef_[0] + modelo.intercept_

plt.plot(X,y_pred, c='r')
```
