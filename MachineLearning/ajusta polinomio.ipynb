Tenemos unos datos que se ajustan bien a un polinomio de grado 2.
Recordamos que un polinomio de grado 2 tiene la forma:
y = ax² + bx + c

Tan solo tendremos que dibujar los datos contenidos en el fichero datos.npz, y dibujar un polinomio ajustando (por ensayo y error) a, b y c.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Cargamos los datos
datos = np.load('datos.npz')
x = datos['x']
y_true = datos['y']
```

```python
#Vemos los datos
plt.scatter(x, y_true)
```

El problema es que la x no está ordenada.
Tendremos que ordenar la x y también la y.

```python
indices = np.argsort(x)

x = x[indices]
y_pred = y_true[indices]
y_true = y_true[indices]
```

```python
#dibujamos de nuevo
plt.scatter(x, y_true)
plt.plot(x, y_pred, c='r')
```
